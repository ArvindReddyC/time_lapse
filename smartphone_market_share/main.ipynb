```python
#================================== packages used ====================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#================================== use of converters to get data ====================================
w = lambda x: (x.replace('%',''))
pdVar = pd.read_html('http://www.counterpointresearch.com/india-smartphone-share/',index_col=0,skiprows=1,header=0,converters={ 1 : w , 2 : w,3 : w,4 : w,5 : w })
```

```python
#================================== Getting data from website====================================
w = lambda x: (x.replace('%',''))
pdVar = pd.read_html('http://www.counterpointresearch.com/india-smartphone-share/',index_col=0,skiprows=1,header=0)

converted_df = []
for each in pdVar:
    converted_df.append( each.applymap(w) )
```

```python
#================================== converting to int, Snapshot of data ====================================
ints = lambda y: ( int(y) )

converted_df[0] = converted_df[0].applymap(ints)
converted_df[0]
```

```python
#================================== bar graphing ====================================
f,ax = plt.subplots(2,3,figsize = (15,10))
colors = plt.cm.Dark2(range(6))
ax[0,0].barh( list(converted_df[0].index) , list(converted_df[0].iloc[:,0]),color=colors )
ax[0,1].barh( list(converted_df[0].index), width =  list(converted_df[0].iloc[:,1]) ,color=colors)
ax[0,2].barh( list(converted_df[0].index) , list(converted_df[0].iloc[:,2]),color=colors )
ax[1,0].barh( list(converted_df[0].index) , list(converted_df[0].iloc[:,3]) ,color=colors)
ax[1,1].barh( list(converted_df[0].index) , list(converted_df[0].iloc[:,4]),color=colors )
ax[1,2].barh( list(converted_df[0].index) , list(converted_df[0].iloc[:,5]) ,color=colors)
```
